## Punto 2: Medidas de Dispersión
En este notebook, exploraremos el concepto de medidas de dispersión utilizando ejemplos y visualizaciones.

## Importar Bibliotecas y Crear Datos

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos con outliers
data = np.concatenate([np.random.normal(0, 1, 90), np.array([8, 10, -12])])
```

## Gráfico de Dispersión

```python
# Crear un gráfico de dispersión
plt.figure(figsize=(8, 6))
plt.scatter(range(len(data)), data, color='blue', label='Datos')
plt.axhline(y=np.mean(data), color='red', linestyle='--', label='Media')
plt.axhline(y=np.median(data), color='green', linestyle='--', label='Mediana')
plt.title('Gráfico de Dispersión con Media y Mediana')
plt.xlabel('Índice del Dato')
plt.ylabel('Valor')
plt.legend()
plt.show()
```

## Explicación
En este ejemplo, hemos creado un conjunto de datos que contiene valores normales junto con algunos valores atípicos (outliers). Luego, hemos representado visualmente estos datos en un gráfico de dispersión.

En el gráfico de dispersión:
- Los puntos azules representan los datos reales.
- La línea roja discontinua representa la media de los datos.
- La línea verde discontinua representa la mediana de los datos.

Observaciones:
- Los valores atípicos (outliers) están claramente separados de la mayoría de los datos.
- La media (línea roja) se ve influenciada por los valores atípicos y se encuentra más hacia la derecha.
- La mediana (línea verde) es menos afectada por los valores atípicos y se encuentra más cerca del centro de la distribución.

Este gráfico nos ilustra cómo los valores atípicos pueden afectar las medidas de tendencia central como la media y la mediana, y cómo la mediana puede ser una medida de tendencia central más robusta en presencia de valores atípicos.
