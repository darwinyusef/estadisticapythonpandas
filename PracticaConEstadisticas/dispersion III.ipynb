## Punto 2: Medidas de Dispersión
En este notebook, exploraremos el concepto de medidas de dispersión utilizando ejemplos y visualizaciones.

## Importar Bibliotecas y Crear Datos

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos con outliers
data = np.concatenate([np.random.normal(0, 1, 60), np.random.normal(3, 1, 20), np.zeros(20)])
```

## Gráfico de Dispersión Ascendente

```python
# Dividir datos en positivos, negativos y nulos
positive_data = data[data > 0]
negative_data = data[data < 0]
zero_data = data[data == 0]

# Crear un gráfico de dispersión ascendente
plt.figure(figsize=(10, 6))
plt.scatter(range(len(positive_data)), positive_data, color='green', label='Positivos')
plt.scatter(range(len(negative_data)), negative_data, color='red', label='Negativos')
plt.scatter(range(len(zero_data)), zero_data, color='blue', label='Nulos')
plt.title('Gráfico de Dispersión')
plt.xlabel('Índice del Dato')
plt.ylabel('Valor')
plt.legend()
plt.show()
```

## Explicación
En este ejemplo, hemos creado un conjunto de datos que contiene valores positivos, negativos y nulos, y luego representamos visualmente estos datos en un gráfico de dispersión ascendente.

En el gráfico de dispersión:
- Los puntos verdes representan los valores positivos.
- Los puntos rojos representan los valores negativos.
- Los puntos azules representan los valores nulos.

Observaciones:
- Los valores nulos (ceros) están claramente separados de los valores positivos y negativos.
- Los valores negativos tienen una mayor dispersión a lo largo del eje y, lo que indica mayor variabilidad.
- Los valores positivos tienen una menor dispersión a lo largo del eje y en comparación con los valores negativos.

Este gráfico ilustra cómo los diferentes grupos de valores (positivos, negativos y nulos) se distribuyen a lo largo del eje y y cómo la dispersión puede variar entre estos grupos.
