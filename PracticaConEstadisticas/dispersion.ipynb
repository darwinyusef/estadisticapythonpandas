# Análisis de Valores Atípicos y Datos Anómalos

En este notebook, exploraremos cómo detectar y visualizar valores atípicos (outliers) en un conjunto de datos, utilizando gráficos de caja (box plots).

## Valores Atípicos y Datos Anómalos

Los valores atípicos son observaciones que se desvían significativamente del patrón general de los datos. Los gráficos de caja (box plots) son una excelente herramienta para visualizar valores atípicos en un conjunto de datos. A continuación, generaremos un gráfico de caja para ilustrar este concepto.

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos con outliers
data = np.concatenate([np.random.normal(0, 1, 90), np.array([8, 10, -12])])

# Crear un gráfico de caja
plt.figure(figsize=(6, 4))
plt.boxplot(data)
plt.title('Gráfico de Caja con Valores Atípicos')
plt.ylabel('Valores')
plt.show()
```

En el gráfico de caja anterior, podemos ver que los valores -12, 8 y 10 son atípicos en comparación con el resto de los datos. Estos valores están fuera del rango intercuartílico y se representan como puntos individuales fuera de las "alas" del gráfico de caja.

## Conclusiones

El análisis de valores atípicos es esencial para comprender mejor la calidad y la integridad de los datos. Los gráficos de caja son una herramienta poderosa para visualizar estos valores y evaluar su impacto en el análisis de datos. Detectar y manejar valores atípicos adecuadamente es crucial para tomar decisiones informadas y construir modelos precisos en ciencia de datos.
